```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout

def dataRename(data):
    data = data[['종가', '매도량', '매수량']]
    data.rename(columns={
        '종가': 'close',
        '매도량': 'sell',
        '매수량': 'buy',
    }, inplace=True)
    return data

# 데이터 로드
data = pd.read_csv('./000660.csv')

df = dataRename(data)

# 정규화
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df)

# 슬라이딩 윈도우 생성
def create_dataset(data, time_step=5):
    X, y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:(i + time_step), :])
        y.append(1 if data[i + time_step, 0] > data[i + time_step - 1, 0] else 0)  # 상승/하락
    return np.array(X), np.array(y)

X, y = create_dataset(data_scaled)

# 모델 설계
def build_model(input_shape):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=input_shape))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.5))  # Dropout 추가
    model.add(Flatten())
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))  # 이진 분류
    return model

# 모델 컴파일 및 학습
input_shape = (X.shape[1], X.shape[2])  # (5, 3)
model = build_model(input_shape)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 학습
model.fit(X, y, epochs=50, batch_size=32, validation_split=0.2)

# 테스트 데이터 준비 (예시)
# X_test, y_test = create_dataset(test_data_scaled)

# 예측 (테스트 데이터 필요)
# predictions = model.predict(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Flatten, Dense, Dropout, Layer
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LayerNormalization
from tensorflow.keras.layers import MultiHeadAttention

class TransformerBlock(Layer):
    def __init__(self, num_heads, key_dim, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attention = MultiHeadAttention(num_heads=num_heads, key_dim=key_dim)
        self.ffn = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dropout(rate),
            Dense(key_dim)  # 출력 차원을 key_dim으로 설정
        ])
        self.layernorm1 = LayerNormalization(epsilon=1e-6)
        self.layernorm2 = LayerNormalization(epsilon=1e-6)
        self.dropout1 = Dropout(rate)
        self.dropout2 = Dropout(rate)

    def call(self, inputs, training):
        attn_output = self.attention(inputs, inputs)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.layernorm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output)

def build_transformer_model(input_shape):
    inputs = Input(shape=input_shape)
    x = Conv1D(filters=64, kernel_size=2, activation='relu')(inputs)
    x = MaxPooling1D(pool_size=2)(x)

    # 차원 조정
    x = tf.expand_dims(x, axis=1)  # (batch_size, 1, features)로 변경
    x = TransformerBlock(num_heads=4, key_dim=64, ff_dim=32)(x)

    x = Flatten()(x)  # 다시 Flatten
    x = Dense(32, activation='relu')(x)
    outputs = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=outputs)  # Model로 변경
    return model



# 모델 컴파일 및 학습
model = build_transformer_model(input_shape)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 학습
model.fit(X, y, epochs=50, batch_size=32, validation_split=0.2)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, Input, LayerNormalization, MultiHeadAttention, GlobalAveragePooling1D

# 데이터 전처리 및 로드 섹션은 이전과 동일하므로 생략합니다.

# 로버스트 스케일러 초기화
scaler = RobustScaler()

# 데이터 스케일링
X_scaled = scaler.fit_transform(X.reshape(-1, X.shape[2])).reshape(X.shape)

# CNN 모델 설계
def build_cnn(input_shape):
    inputs = Input(shape=input_shape)
    x = Conv1D(filters=64, kernel_size=2, activation='relu')(inputs)
    x = MaxPooling1D(pool_size=2)(x)
    x = Flatten()(x)
    return Model(inputs, x)

# CNN 모델 생성
input_shape = (X_scaled.shape[1], X_scaled.shape[2])  # (5, 3)
cnn_model = build_cnn(input_shape)

# CNN 특징 추출
cnn_output = cnn_model.predict(X_scaled)

# 트랜스포머 모델 설계
def build_transformer(input_shape):
    inputs = Input(shape=input_shape)
    x = MultiHeadAttention(num_heads=4, key_dim=2)(inputs, inputs)
    x = LayerNormalization(epsilon=1e-6)(x)
    x = Dropout(0.1)(x)
    x = GlobalAveragePooling1D()(x)
    x = Dense(32, activation='relu')(x)
    outputs = Dense(1, activation='sigmoid')(x)  # 이진 분류
    return Model(inputs, outputs)

# CNN 출력을 트랜스포머 입력으로 재구성
cnn_output = cnn_output.reshape(cnn_output.shape[0], cnn_output.shape[1], 1)  # (샘플 수, 특징 수, 1)

# 트랜스포머 모델 생성
transformer_model = build_transformer((cnn_output.shape[1], 1))

# 모델 컴파일
transformer_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 트랜스포머 모델 학습
transformer_model.fit(cnn_output, y, epochs=50, batch_size=32, validation_split=0.2)
```
