```python
# RSI 계산 함수
def calculate_rsi(data, period=14):
    delta = data.diff()  # 가격 변화량 계산
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()  # 평균 상승폭
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()  # 평균 하락폭
    
    rs = gain / loss  # RS 계산
    rsi = 100 - (100 / (1 + rs))  # RSI 계산
    
    return rsi

# RSI 컬럼 추가
df['RSI'] = calculate_rsi(df['종가'], period=14)
```
